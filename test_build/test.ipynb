```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load a 3d model with obj file and plot it
def load_obj(file_name):
    with open(file_name, 'r') as f:
        vertices = []
        faces = []
        for line in f:
            if line.startswith('#'):
                continue
            values = line.split()
            if not values:
                continue
            if values[0] == 'v':
                v = [float(i) for i in values[1:4]]
                vertices.append(v)
            elif values[0] == 'f':
                face = [int(i.split('/')[0]) for i in values[1:4]]
                faces.append(face)
    return np.array(vertices), np.array(faces)

# show the 3d model
def show_model(vertices, faces):
    x = vertices[:, 0]
    y = vertices[:, 1]
    z = vertices[:, 2]
    ax = plt.axes(projection='3d')
    ax.scatter3D(x, y, z, cmap='Greens')
    ax.plot_trisurf(x, y, z, triangles=faces, cmap='viridis', edgecolor='none')
    plt.show()
    
# load the 3d model
vertices, faces = load_obj('3d-lowpolyf1/f1 low poly.obj')
# show the 3d model
show_model(vertices, faces)
```
